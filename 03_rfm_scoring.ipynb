```python
import pandas as pd
from sqlalchemy import create_engine, text

engine = create_engine('sqlite:///data/rfm.db')
print('Connected to database ✓')
```

```python
with open('sql/01_rfm_scoring.sql', 'r') as f:
    sql = f.read()

df_rfm = pd.read_sql(sql, engine)
print(f'RFM scores calculated for {len(df_rfm):,} customers')
df_rfm.head()
```

```python
df_rfm.to_sql('rfm_scores', engine, if_exists='replace', index=False)
print('rfm_scores table saved to SQLite ✓')
```

```python
with engine.connect() as conn:
    count = conn.execute(text('SELECT COUNT(*) FROM rfm_scores')).fetchone()[0]

print(f'rfm_scores table: {count:,} customers')
```
